```python
def f2(x):
    return 4 / (1 + x * x)


def _calcu_integral():
    left_riemann = 0.0
    right_riemann = 0.0
    midpoint_riemann = 0.0
    a = 0.0
    b = 1.0
    h = b - a
    for n in range(1, 8192):
        n *= 2
        for i in range(1, n):
            left_riemann += f2(a + h * (i - 1))
            right_riemann += f2(a + h * i)
            midpoint_riemann += f2(a + h * (i - 0.5))
        left_riemann *= h
        right_riemann *= h
        midpoint_riemann *= h
        print(f"left_riemann = {format(left_riemann,'.3f')}, right_riemann = {format(right_riemann,'.3f')},midpoint_riemann = {format(midpoint_riemann,'.3f')}")
        h /= 2.0
```

```python
_calcu_integral()
```
