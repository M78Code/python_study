```python
# 时长分布停机

# 1.创建画布
plt.figure(figsize=(20,8),dpi=100)

# 2. 绘制
plt.hist(movie['Runtime (Minutes)'].values,20)

# 2.1 增加x轴刻度
max_ = movie['Runtime (Minutes)'].max()
min_ = movie['Runtime (Minutes)'].min()

x1 = np.linspace(min_, max_, 21)
# print(x1)

plt.xticks(x1)

plt.grid()

# 3. 显示
plt.show()
```

# 问题3： 
对于一组电影数据，如果希望统计电影分类(genre)的情况，应该如何处理数据？

```python
# 思路分析
  # 1，创建一个全为0的dataFrame，列索引置为电影的分类，temp_df
  # 2，遍历每一部电影，temp_df中把分类出现的列的值置为1
  # 3，求和

movie.head()
```

```python
movie['Genre'].head()
```

```python
[i for i in movie['Genre']]
```

```python
# 进行字符串分割
temp_list = [i.split(',') for i in movie['Genre']]
temp_list
```

```python
[j for i in temp_list for j in i]
```

```python
array_list = np.unique([j for i in temp_list for j in i])
array_list
```

```python
array_list.shape[0]
```

```python
movie.shape
```

```python
array_list.shape
```

```python
np.zeros([movie.shape[0],array_list.shape[0]])
```

```python
# 创建一个全为零的df
genera_zero = pd.DataFrame(np.zeros([movie.shape[0],array_list.shape[0]]),columns=array_list)
```

```python
genera_zero.head()
```

```python
for i in range(1000):
    genera_zero.loc[i, temp_list[i]] = 1
```

```python
genera_zero.sum()
```

```python
genera_zero.sum().sort_values(ascending = False).plot(kind='bar',figsize=(20,8),)
```

```python
genera_zero.sum().sort_values(ascending = False).plot(kind='bar',figsize=(20,8),colormap='cool',fontsize=18)
```
